```python
import pandas as pd
qa_data_path = "/data_x/guro/guro_eval.jsonl"
qa_data = pd.read_json(qa_data_path, lines=True)
```

```python
qa_data
```

```python
qa = []

for index, data in enumerate(qa_data["query"]):
    qa.append({
        "qid": str(index),
        "query": data,
        "generation_gt": [qa_data["answer"][index]],
        "retrieval_gt": [["None"]]
    })
```

```python
save_dir = "/data_x/guro/DATA/ASCO_RAW"

qa_df = pd.DataFrame(qa)
qa_df.to_parquet(save_dir + "/qa.parquet")
```
